It is vital data is stationary, otherwise, we will not have a reliable basis to build the forecasting on. The presence of a data trend implies the data is non-stationary. A Dickey-Fuller test will be conducted to check for stationarity.

### 4. Stationarity Test 
##### a. Dickey-Fuller Test

```python
# The adfuller() function imports from the model from statsmodels; but running it will only return an array of numbers
# This is the reason to define a function that prints the correct output from that array

# Importing the adfuller() function

from statsmodels.tsa.stattools import adfuller 

# Performing the Dickey-Fuller test:

def dickey_fuller(timeseries): # Define the function
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the test using the function on the time series
dickey_fuller(data['EUR Price'])
```

H0 = Presence of a unit root (non-stationary)
HA = Unit root present (stationary)
The test statistic of  -2.920929 is smaller than the critical value (5%) of -2.900395, meaning the null hypothesis can be rejected. The result is also significant as the p-value is smaller than 0.05.

##### b. Autocorrelation Test

```python
# Check out a plot of autocorrelations

from statsmodels.graphics.tsaplots import acf, plot_acf, plot_pacf # Here, you import the autocorrelation and partial correlation plots

plot_acf(df_sub, lags=30)
plt.show()
```

The test shows that there are 5 vertical lines that go above the blue edge of the confidence interval, indicating that the lags are significantly correlated with each other and the data is non-stationary.

### 6. Stationarizing Data

```python
# Conducting the first round of differencing by subtracting the previous data point from the current data point
# The df.shift(1) function turns the observation to t-1, making the whole data t - (t -1)

data_diff = df_sub - df_sub.shift(1)
```

```python
# Removing missing values as a result of differencing

data_diff.dropna(inplace = True)
```

```python
# Checking values and headers

data_diff.head()
```

```python
# Checking out what the differencing did to the time-series curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
```

```python
# Re-running the Dickey-Fuller Test to check for stationarity

dickey_fuller(data_diff)
```

```python
# Plotting the autocorrelation again

plot_acf(data_diff)
plt.show()
```

The autocorrelation plot has improved. The stationary is strong due to no vertical lines going above the confidence interval.

### B. Time Series Forecasting in Python Bonus
#### 1. Defining Your Parameters

This is the easiest part of the model’s setup. Stands for the d number of times we needed to difference our data in order to make it stationary. In this case, we differenced one time, which resulted in weak stationarity as determined by the Dickey-Fuller
Test results and autocorrelations plot.
d=1

```python
# Plot autocorrelations to determine parameter q

plot_acf(data_diff)
plt.figure(figsize=(15,5), dpi=100)
plt.show()
```

We have 3 significant autocorrelations here (those that run beyond the blue border), but for
now, we can take a more conservative approach and see what the output is like at q = 3 . It’s
advisable to start with fewer terms as opposed to more and adjust accordingly further down the
line as we begin testing the model.
Despite the ACF and PACF plots giving us some direction into the parameters setup for
ARIMA, there’s no set-in-stone approach for what p and q should be. Modeling i s a matter of
iteration and testing out different parameters until we get a satisfactory result. Therefore, the
starting figures often feel very arbitrary.

```python
# Plot partial autocorrelations to determine parameter p

plot_pacf(data_diff)
plt.figure(figsize=(15,15), dpi=300)
plt.show()
```

Is the parameter responsible for the autoregressive part p of the model
(AR). In the figure above, we can see that we have some highly positive bars. These indicate
significant correlations. Let’s start by picking 4 AR terms to compensate for the highest of the
partial autocorrelations. This makes p = 3.

#### 2. Splitting the Data

Splitting time series data isn’t like splitting regular data sets. This is because we only have one
value per data entry, and in order to test the time prediction of the model, we need consecutive
data. For this reason, we need to split the data by time. In this case, you can split the data right
after the effect of the global financial crisis started wearing off around 2010/2011 and check how
accurately the model predicts how quickly the world economy got back on its feet in the
following quarters.


```python
train = df_sub[df_sub.index <= '2011-01-01'] 
test = df_sub[df_sub.index >= '2011-01-01']
test.shape
```

```python
# Plot both sets in order to see the split

plt.figure(figsize=(15,5), dpi=100)
plt.plot(train)
plt.plot(test)
```

```python
test.shape
```

Before looking at the results, let’s summarize what you need to be looking for in the results of
your model:  
● Model summary shows results of AR and MA terms are statistically significant.  
● On a plot, the curve for the forecast overlaps with the curve for the actual values. Or,
the actual values curve should fall within the forecasted confidence interval.

#### 3. Running and Fitting the Model

p = 3 d = 1 q = 3
model = ARIMA(train, order=(p,d,q))

```python
from statsmodels.tsa.api import ARIMA, SARIMAX # Import the model you need
```

```python
model = ARIMA(train, order=(3,1,3))  
fitted = model.fit()  
print(fitted.summary()) # Check model summary
```

In the box on the left, we can see some coefficients that start with “ar” and “ma.” These are
the terms we added when setting the model parameters. In the box on the right, we can
see whether those coefficients are statistically significant (p <= 0.05). 

Having significant coefficients is good news for sure, but we still need to plot the curve of the
forecasted values against the test values in order to get validation for the fit of our model. If the
model is, indeed, a good fit, the forecast line should overlap with the actual line. Let’s take a
look:

```python
# Forecast
fc = fitted.get_forecast(10, alpha=0.05)  # 95% confidence interval  --> alpha = 0.05; 10 is the shape of the test set

# Make into pandas series
fc_series = pd.Series(fc.predicted_mean, index=test.index)
lower_series = pd.Series(fc.conf_int().iloc[:, 0], index=test.index)
upper_series = pd.Series(fc.conf_int().iloc[:, 1], index=test.index)
```

```python
# Plot
plt.figure(figsize=(16,4), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.2)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

#### 4. Testing for stationarity

```python
# The adfuller() function will import from the model from statsmodels for the test; however, running it will only return 
# an array of numbers. This is why you need to also define a function that prints the correct output from that array.

from statsmodels.tsa.stattools import adfuller # Import the adfuller() function

def dickey_fuller(timeseries): # Define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the test using the function on the time series
dickey_fuller(df_sub['EUR Price'])
```

H0 = Unit root present (non-stationary)
HA = Unit root not present (stationary)
- If test statistic larger than critical value = can’t reject null hypothesis.
- If test statistic smaller than critical value = can reject null hypothesis

??????????Using significance level of 5%, the critical Value is -2.933297. While, the test statistic is  -3.429249, which shows that it is SMALLER than critical value. This means that the null hypothesis can be rejected. In other words, there is a nounit root in our data and our data is stationary.
Using p-value, it also shows the same result. P-value ( 0.010001) is also SMALLER than 0.05.

```python
# Check out a plot of autocorrelations

from statsmodels.graphics.tsaplots import acf, plot_acf, plot_pacf # Here, you import the autocorrelation and partial correlation plots

plot_acf(df_sub, lags=30)
plt.show()
```

The test shows that there are vertical lines that go above the blue edge of the confidence interval, indicating that the lags are significantly correlated with each other and the data is non-stationary.
