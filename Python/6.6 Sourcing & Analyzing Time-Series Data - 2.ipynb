#### 4. Testing for stationarity

```python
# The adfuller() function will import from the model from statsmodels for the test; however, running it will only return 
# an array of numbers. This is why you need to also define a function that prints the correct output from that array.

from statsmodels.tsa.stattools import adfuller # Import the adfuller() function

def dickey_fuller(timeseries): # Define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the test using the function on the time series
dickey_fuller(df_sub['EUR Price'])
```

H0 = Unit root present (non-stationary)
HA = Unit root not present (stationary)
- If test statistic larger than critical value = can’t reject null hypothesis.
- If test statistic smaller than critical value = can reject null hypothesis

Using significance level of 5%, the critical Value is -2.957219. While, the test statistic is  -2.197290, which shows that it is larger than critical value. This means that the null hypothesis cannot be rejected. In other words, there is a unit root in our data and our data is non-stationary.
Using p-value, it also shows the same result. P-value ( 0.2072307) is also larger than 0.05.

```python
# Check out a plot of autocorrelations

from statsmodels.graphics.tsaplots import acf, plot_acf, plot_pacf # Here, you import the autocorrelation and partial correlation plots

plot_acf(df_sub, lags=30)
plt.show()
```

The test shows that there are vertical lines that go above the blue edge of the confidence interval, indicating that the lags are significantly correlated with each other and the data is non-stationary.

#### 05. Stationarizing Data

```python
df_diff = df_sub - df_sub.shift(1) # The df.shift(1) function turns the observation to t-1, making the whole thing t - (t -1)
```

```python
df_diff.dropna(inplace = True) # Here, you remove the missing values that came about as a result of the differencing. 
# You need to remove these or you won't be able to run the Dickey-Fuller test.
```

```python
df_diff.head()
```

```python
# Check out what the differencing did to the time-series curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(df_diff)
```

```python
# Re-running the Dickey-Fuller Test to check for stationarity

dickey_fuller(df_diff)
```

The results show that the data is now stationary. The test statistic value (-6.199769e+000) is less than the critical value (5%) of   -2.964071e+00 and the p-value is smaller than 0.05. Therefore, the test result is significant and the null hypothesis can be rejected.

```python
# Plotting the autocorrelation again

plot_acf(df_diff)
plt.show()
```

The autocorrelation plot has improved. The stationary is strong due to no vertical lines going above the confidence interval.

#### 2. Splitting the Data

```python
# Split the data

train = df_diff[df_diff.index <= '2018-12-31'] 
test = df_diff[df_diff.index >= '2018-12-31']
```

```python
# Plot both sets in order to see the split

plt.figure(figsize=(15,5), dpi=100)
plt.plot(train)
plt.plot(test)
```

```python
test.shape
```

### 3. Running and Fitting the Model

```python
from statsmodels.tsa.api import ARIMA, SARIMAX # Import the model you need
```

```python
model = ARIMA(train, order=(4, 1, 4))  
fitted = model.fit()
print(fitted.summary()) # Check model summary
```

```python
# Forecast
fc = fitted.get_forecast(33, alpha=0.05)  # 95% confidence interval  --> alpha = 0.05; 33 is the shape of the test set

# Make into pandas series
fc_series = pd.Series(fc.predicted_mean, index=test.index)
lower_series = pd.Series(fc.conf_int().iloc[:, 0], index=test.index)
upper_series = pd.Series(fc.conf_int().iloc[:, 1], index=test.index)
```

```python
# Plot
plt.figure(figsize=(12,4), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.2)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```
